```python
import numpy
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

text = "I want to start working as a programmer. I'm learning programming at Tel-Ran course.\
The teacher says to us that for becoming a programmer you should perform Homeworks.\
    But because of family and work I don't have time to do Homeworks"
text_arr: list[str] = text.split(".")
cleaned_text = [res for s in text_arr if (res := s.strip())]
# query = ["What do you want?"]
# query = ["Where are you learning programming?"]
# query =["What does the teacher say to you?"]
query = ["Why don't you do Homeworks?"]
vectorizer = TfidfVectorizer()
embeddings = vectorizer.fit_transform(cleaned_text + query).toarray()
embeddings_text = embeddings[:-1]
embeddings_query = embeddings[-1:]
similarities = cosine_similarity(embeddings_text, embeddings_query)
best_ind = numpy.argmax(similarities)
cleaned_text[best_ind]
```
