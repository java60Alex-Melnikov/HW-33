```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
model = YOLO("yolov8m.pt")
results = model('street.jpg')
img_annotated = results[0].plot()
plt.figure(figsize=(10,8))
plt.imshow(img_annotated)
plt.title("Detection Objects")
plt.axis("off")
plt.show()
```

```python
model = YOLO("yolov8m-seg.pt")
results = model('street.jpg')
img_annotated = results[0].plot()
plt.figure(figsize=(8,10))
plt.imshow(img_annotated)
plt.title("Segmentation Objects")
plt.axis("off")
plt.show()
```

```python
model = YOLO("yolov8l-cls.pt")
results = model('street.jpg')
img_annotated = results[0].plot()
plt.figure(figsize=(5,5))
plt.imshow(img_annotated)
plt.title("Image Classification")
plt.axis("off")
plt.show()
```

```python
import pandas as pd
model = YOLO("yolov8m-seg.pt")
results = model('street.jpg')
boxes = results[0].boxes
xyxy = boxes.xyxy.cpu().numpy()
cls = boxes.cls.cpu().numpy()
conf = boxes.conf.cpu().numpy()
allNames = results[0].names
names = [allNames[c] for c in cls]
df = pd.DataFrame(xyxy, columns=["xmin", "ymin", "xmax", "ymax"])
df["name"] = names
df["confidence"] = conf
df
```
